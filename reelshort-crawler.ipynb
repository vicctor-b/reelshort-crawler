```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time

# Configurar o Selenium
options = Options()
options.add_argument('--headless')
options.add_argument('--disable-gpu')
driver = webdriver.Chrome(options=options)

base_url = "https://www.reelshort.com/pt/full-episodes/a-vida-secreta-do-meu-marido-bilion%C3%A1rio-68196320aeae6784260a959a/{}"
hrefs = []

for page in range(1, 8):
    url = base_url.format(page)
    print(f"Navegando para {url}")
    driver.get(url)
    time.sleep(3)  # espera o JS carregar
    
    # Pega o HTML após renderização
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Seleciona a div principal
    container = soup.select_one('div.border-t-1px.border-t-\\[\\#FFFFFF33\\].border-t-solid')
    if not container:
        print(f"Nenhum conteúdo encontrado na página {page}")
        continue

    for child_div in container.find_all(recursive=False):
        first_inner_div = child_div.find('div')
        if first_inner_div:
            first_a = first_inner_div.find('a', href=True)
            if first_a:
                hrefs.append(first_a['href'])

driver.quit()

# Exibir os hrefs encontrados
for h in hrefs:
    print(h)
```

```python
import ffmpeg
import os
from concurrent.futures import ThreadPoolExecutor, as_completed

# Diretório temporário para arquivos intermediários
temp_dir = "temp_videos"
os.makedirs(temp_dir, exist_ok=True)

# Função para processar um único link .m3u8
def process_m3u8(link, index):
    output_file = os.path.join(temp_dir, f"temp_{index}.mp4")
    try:
        # Baixa e converte o .m3u8 para .mp4
        stream = ffmpeg.input(link)
        stream = ffmpeg.output(stream, output_file, c='copy', loglevel='error')
        ffmpeg.run(stream)
        print(f"Baixado e convertido: {output_file}")
        return output_file
    except ffmpeg.Error as e:
        print(f"Erro ao processar {link}: {e.stderr.decode()}")
        return None

# Passo 1: Processar links em paralelo com até 8 threads
temp_files = []
with ThreadPoolExecutor(max_workers=8) as executor:
    # Submeter todas as tarefas
    future_to_index = {executor.submit(process_m3u8, link, i): i for i, link in enumerate(m3u8_links)}
    # Coletar resultados conforme completam
    for future in as_completed(future_to_index):
        result = future.result()
        if result:
            temp_files.append(result)

# Ordenar arquivos temporários por índice para manter a ordem original
temp_files.sort(key=lambda x: int(x.split('_')[-1].split('.')[0]))

# Passo 2: Criar arquivo de concatenação para o ffmpeg
concat_list_path = os.path.join(temp_dir, "concat_list.txt")
with open(concat_list_path, 'w') as f:
    for temp_file in temp_files:
        f.write(f"file '{temp_file}'\n")
```

```python
# Passo 3: Concatenar os vídeos em um único .mp4
output_final = "output_final.mp4"
try:
    stream = ffmpeg.input(concat_list_path, format='concat', safe=0)
    stream = ffmpeg.output(stream, output_final, c='copy', loglevel='error')
    out, err = ffmpeg.run(stream, capture_stdout=True, capture_stderr=True)
    print(f"Vídeo final criado: {output_final}")
except ffmpeg.Error as e:
    stderr = e.stderr.decode('utf-8') if e.stderr else "No stderr output"
    stdout = e.stdout.decode('utf-8') if e.stdout else "No stdout output"
    print(f"Erro ao concatenar vídeos:\nSTDERR: {stderr}\nSTDOUT: {stdout}")
    exit(1)
```

```python
# Passo 4: Limpar arquivos temporários
for temp_file in temp_files:
    os.remove(temp_file)
os.remove(concat_list_path)
os.rmdir(temp_dir)
print("Arquivos temporários removidos.")
```
